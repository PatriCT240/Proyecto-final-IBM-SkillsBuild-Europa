```python
## Cambia el nombre de la columna.
df = df.rename(columns={"availability 365": "days_booked"})

print(df.columns)
```

```python
## Convierte todos los nombres de columna a minúsculas y sustituye los espacios por un guión bajo "_".
# Convierto todos los nombres de columna a minúsculas
df.columns = df.columns.str.lower()

# Sustituyo los espacios por un guión bajo "_"
df.columns = df.columns.str.replace(" ", "_")

print(df.columns)
```

```python
## Elimina el signo de dólares y la coma de las columnas. Si es necesario, convierte estas dos columnas al tipo de datos adecuado.
# Elimino el signo del dólar y la coma de las columnas.
df["price"] = df["price"].str.replace(r"[$,]", "", regex=True)
df["service_fee"] = df["service_fee"].str.replace(r"[$,]", "", regex=True)

# Convierto las columnas a numéricas
df["price"] = df["price"].astype("float64")
df["service_fee"] = df["service_fee"].astype("float64")

df.info()
```

```python
#Durante el ejercicio he encontrado que las ciudades Manhattan y Brooklyn estan escritas de manera diferente, las unifico
df["neighbourhood_group"] = df["neighbourhood_group"].replace({"manhatan": "Manhattan", "brookln": "Brooklyn"})
```

```python
#Para ver mejor los gráficos
sns.set_context("notebook")
sns.set_theme(style="ticks")
sns.set_palette("pastel")
plt.rcParams["figure.figsize"]=9,4
```

```python
## Enumerar los distintos tipos de habitaciones disponibles en Airnb
# Obtengo los tipos de habitaciones disponibles
tipo_hab = df["room_type"].unique()

# Asigno color a cada tipo de habitación
colores = {
    "Entire home/apt": "#b2ebf2",
    "Private room": "#e6f5d0",
    "Shared room": "#f5d0e6",
    "Hotel room": "#d8d8d8"
}

# Creo un gráfico de barras
plt.bar(tipo_hab, df["room_type"].value_counts(), color=colores.values())

# Mejoras gráfico
plt.title("Tipos de habitaciones disponibles en Airbnb")
plt.xlabel("Tipo de habitación")
plt.ylabel("Número de anuncios")

for i, (tipo_hab, valor) in enumerate(zip(tipo_hab, df["room_type"].value_counts())):
    plt.annotate(f"{valor:,}", (i, valor), ha="center", va="bottom")
    
plt.show()
```

```python
## Qué tipo de habitación se adhiere a una política de cancelación más estricta
hab_canc = pd.crosstab(df["cancellation_policy"], df["room_type"])

cmap = sns.color_palette(["lightgreen", "gold", "salmon"])

ax = sns.heatmap(hab_canc, annot=True, cmap= cmap, fmt='d')
ax.set_yticklabels(ax.get_yticklabels(), rotation=0, ha="right")
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha="right")
ax.set_xlabel("Tipo de habitación")
ax.set_ylabel("Política de cancelación")
ax.set_title("Política de cancelación según el tipo de habitación")

plt.show()
```

```python
## Enumere los precios por grupo de barrios y mencione también cuál es el grupo de barrios más caro para los alquileres
# Calculo el precio promedio por grupo de barrios
precio_barrio = df.groupby("neighbourhood_group")["price"].mean().sort_values(ascending=False)

# Lista de los precios por grupo de barrios
precio_barrio_list = precio_barrio.reset_index()
precio_barrio_list["price"] = precio_barrio_list["price"].round(2)

print("Los precios promedios según grupo de barrios son:")
print(precio_barrio_list[["neighbourhood_group", "price"]])

# Grupo de barrios más caro
barrio_mas_caro = precio_barrio.idxmax()
precio_mas_caro = precio_barrio.max()

print(f"\nEl grupo de barrios más caro es '{barrio_mas_caro}' con un precio promedio de ${precio_mas_caro:.2f}")
```

```python
## Enumere los 10 barrios más caros por orden creciente de precio con la ayuda de un gráfico de barras horizontales. 
# Calculo el precio promedio por barrio
precio_bar = df.groupby("neighbourhood")["price"].mean().sort_values(ascending=False)

# Los 10 barrios más caros
top_10_caros = precio_bar.head(10)

# Lista barrios más caros
top_10_caros_list = top_10_caros.reset_index()

# Gráfico de barras horizontales
plt.figure(figsize=(10, 6))
sns.barplot(x="price", y="neighbourhood", data=top_10_caros_list, palette="viridis")
plt.title("Los 10 Barrios más caros")
plt.xlabel("Precio Promedio")
plt.ylabel("Barrio")

for index, value in enumerate(top_10_caros_list["price"]):
    plt.text(value, index, f'${value:.2f}', color='black', ha="right", va="center")

plt.show()

# Barrio más barato
barrio_mas_barato = precio_bar.idxmin()

print(f"\nEl barrio más barato es: {barrio_mas_barato} con un precio promedio de ${precio_bar.min():.2f}")
```

```python
## Enumere los barrios que ofrecen alquileres a corto plazo de menos de 10 días. Ilustrar con un gráfico de barras
# Alquileres a corto plazo de menos de 10 días
alq_corto_plazo = df[df["minimum_nights"] < 10]

# Frecuencia de cada barrio
frec_barrio = alq_corto_plazo["neighbourhood"].value_counts()

# Selecciono los 10 barrios con más alquileres a corto plazo (porque la lista puede ser muy larga)
top_10_barrios = frec_barrio.head(10)

# Gráfico de barras
plt.figure(figsize=(12, 6))
barplot = sns.barplot(x=top_10_barrios.index, y=top_10_barrios.values, palette="pastel")
plt.title("Barrios con alquileres a corto plazo (menos de 10 días)")
plt.xlabel("Barrio")
plt.ylabel("Número de alquileres a corto plazo")
plt.xticks(rotation=45, ha="right")

for index, value in enumerate(top_10_barrios.values):
    barplot.text(index, value + 1, str(value), color='black', ha="center")

plt.show()
```

```python
## Enumere los precios con respecto al tipo de habitación utilizando un gráfico de barras y exponga también sus inferencias
plt.figure(figsize=(10, 6))
boxplot = sns.boxplot(x="room_type", y="price", data=df, palette="Set2")
plt.title("Precios según tipo de habitación")
plt.xlabel("Tipo de habitación")
plt.ylabel("Precio ($)")

# Añado las medianas
medianas = df.groupby("room_type")["price"].median().reset_index()

for index, row in medianas.iterrows():
    plt.annotate(f"Mediana: ${row['price']:.2f}", 
                 xy=(index, row['price']),
                 xytext=(0,10), 
                 textcoords="offset points",
                 ha="center", va="center", color="black", size=10)

plt.xticks(rotation=45, ha="right")
plt.show()
```

```python
## Cree un gráfico circular que muestre la distribución de los días reservados para cada grupo de barrios
#Días reservados
dias_reser = df[df["days_booked"] > 0]

#Grupos de barrios con días reservados
barrio_reser = dias_reser["neighbourhood_group"]

#Frecuencia de cada grupo de barrios
barrio_reser_dias = barrio_reser.value_counts()

#Gráfico circular
plt.figure(figsize=(6, 5))
plt.pie(barrio_reser_dias.values, labels=None, autopct='%1.2f%%', startangle=140)
plt.legend(barrio_reser_dias.index, loc="center left", bbox_to_anchor=(1, 0.5), title="Grupo de barrios")
plt.title("Distribución de los días reservados para cada grupo de barrios")
plt.show()
```

```python
## ¿El precio del servicio y el precio de la habitación tienen un impacto mutuo? Ilustre esta relación con un gráfico de 
## dispersión e indique sus inferencias
sns.scatterplot(x="service_fee", y="price", data=df)

plt.title("Relación entre el Precio del servicio y el Precio de la habitación")
plt.xlabel("Precio del servicio ($)")
plt.ylabel("Precio de la habitación ($)")
plt.show()
```

En el gráfico podemos ver como existe una **relación positiva perfecta** entre el precio de la habitación y el precio del servicio. Por tanto, existe una relación lineal entre las dos variables, de modo que un aumento en el precio de la habitación siempre está acompañado por un aumento en el precio del servicio en una proporción constante.

```python
## Utilizando un gráfico lineal muestre en qué año tuvo lugar la máxima construcción de habitaciones
# Agrupo por año y número de construcciones de habitaciones en cada año
const_año = df["construction_year"].value_counts().reset_index()
const_año.columns = ["año", "construcciones"]
const_año["año"] = const_año["año"].astype(int)

# Año con el máximo número de construcciones
año_max_const = const_año.loc[const_año["construcciones"].idxmax(), "año"]

# Gráfico 
plt.figure(figsize=(10, 6))
sns.lineplot(x="año", y="construcciones", data=const_año, marker="o", color="skyblue")

plt.title("Número de construcciones de habitaciones por año")
plt.xlabel("Año de la construcción")
plt.ylabel("Número de construcciones")

# Indico el año con la máxima construcción
plt.annotate(f"Máxima construcción\nen {año_max_const}", 
             xy=(año_max_const, const_año["construcciones"].max()), 
             xytext=(año_max_const + 2, const_año["construcciones"].max() - 50),
             arrowprops=dict(facecolor="salmon", shrink=0.1))

plt.text(año_max_const, const_año["construcciones"].max(), const_año["construcciones"].max(), ha="center", va="bottom", size=10)

plt.xticks(rotation=45, ha="right", ticks=const_año["año"].unique())
plt.show()
```

* Con la ayuda de gráficos de caja ilustra lo siguiente
 * Efecto del número de tasa de revisión en el precio

Si utiliza Python para este ejercicio, por favor incluya el código en las celdas de abajo. Si utiliza cualquier otra herramienta, por favor incluya capturas de pantalla de su trabajo.

```python
## Efecto del número de tasa de revisión en el precio
# Filtro para eliminar valores nulos
df_fil = df.dropna(subset=["review_rate_number", "price"])

# Gráfico
plt.figure(figsize=(10, 6))
sns.boxplot(x="review_rate_number", y="price", data=df_fil)

# Añado las medianas
medianas = df.groupby("review_rate_number")["price"].median().reset_index()

for index, row in medianas.iterrows():
    plt.annotate(f"Mediana: ${row['price']}", 
                 xy=(index, row['price']),
                 xytext=(0,10), 
                 textcoords="offset points",
                 ha="center", va="center", color="black", size=10)

plt.title("Efecto del número de tasa de revisión en el precio")
plt.xlabel("Número de tasa de revisión")
plt.ylabel("Precio")
plt.show()
```

A medida que aumenta la tasa de revisión, el precio va disminuyendo.

```python
## Efecto de la identidad del host verificada en el precio
plt.figure(figsize=(10, 8))
sns.boxplot(x="host_identity_verified", y="price", data=df)

# Añado las medianas
medianas = df.groupby("host_identity_verified")["price"].median().reset_index()

for index, row in medianas.iterrows():
    plt.annotate(f"Mediana: ${row['price']}", 
                 xy=(index, row['price']),
                 xytext=(0,10), 
                 textcoords="offset points",
                 ha="center", va="center", color="black", size=12)

plt.title("Efecto de la identidad del host verificada en el precio")
plt.xlabel("Identidad del host verificada")
plt.ylabel("Precio")

plt.show()
```
